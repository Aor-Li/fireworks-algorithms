```python
import os
import math
import numpy as np
import pickle as pkl
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
func_id = 15
dir1 = 'logs/{}_{}_{}D_func_{}'.format('BBFWA', 'CEC17', 30, func_id)
dir2 = 'logs/{}_{}_{}D_func_{}'.format('round_SRBBFWA', 'CEC17', 30, func_id)

plt.figure(figsize=(20, 10))
for filename in os.listdir(dir2):
    with open(os.path.join(dir2, filename), 'rb') as f:
        data = pkl.load(f)
    y = [math.log(_) for _ in data['traj']]
    x = list(range(len(y)))
    plt.plot(x, y, 'r')

for filename in os.listdir(dir1):
    with open(os.path.join(dir1, filename), 'rb') as f:
        data = pkl.load(f)
    y = [math.log(_) for _ in data['traj']]
    x = list(range(len(y)))
    plt.plot(x, y, 'g')
    
```
